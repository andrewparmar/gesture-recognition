```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np
import scipy.signal as sp
import scipy
```

# Load Image

```python
img = cv2.imread('Shift0.png')

print(img.shape)

# plt.figure(figsize = (250, 350))

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

### Sobel Derivate X Direction

```python
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3, scale=(1/8))
```

### Sobel Derivate Y Direction


```python
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3, scale=(1/8))
```

### Draw

```python
plt.imshow(sobel_x)
# plt.imshow(cv2.cvtColor(sobel_x, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(sobel_y)
# plt.imshow(cv2.cvtColor(sobel_x, cv2.COLOR_BGR2RGB))
plt.show()
```
