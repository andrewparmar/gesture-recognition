```python
%matplotlib inline
import matplotlib.pyplot as plt
import cv2
import numpy as np


# For Your Eyes Only
frizzy = cv2.imread('images/frizzy.png')
froomer = cv2.imread('images/froomer.png')
plt.imshow(cv2.cvtColor(frizzy, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(froomer, cv2.COLOR_BGR2RGB))

frizzy = cv2.cvtColor(frizzy, cv2.COLOR_BGR2GRAY)
froomer = cv2.cvtColor(froomer, cv2.COLOR_BGR2GRAY)
plt.imshow(frizzy, cmap='gray')
plt.show()
plt.imshow(froomer, cmap='gray')
```

```python
# TODO: Find edges in frizzy and froomer images

# Frizzy
# gx_frizzy = cv2.Sobel(frizzy, -1, dx=1, dy=0)
# gy_frizzy = cv2.Sobel(frizzy, -1, dx=0, dy=1)
gx_frizzy = cv2.Canny(frizzy, 100, 200)
gy_frizzy = cv2.Canny(frizzy, 100, 200)

plt.imshow(gx_frizzy, cmap='gray')
plt.show()
plt.imshow(gy_frizzy, cmap='gray')
plt.show()

gmag_frizzy = np.sqrt(gx_frizzy**2 + gy_frizzy**2)
plt.imshow((gmag_frizzy / (4 * np.sqrt(2))).astype(np.uint8), cmap='gray')
plt.show()

# froomer
gx_froomer = cv2.Canny(froomer, 100, 200)
gy_froomer = cv2.Canny(froomer, 100, 200)

plt.imshow(gx_froomer, cmap='gray')
plt.show()
plt.imshow(gy_froomer, cmap='gray')
plt.show()

gmag_froomer = np.sqrt(gx_froomer**2 + gy_froomer**2)
plt.imshow((gmag_froomer / (4 * np.sqrt(2))).astype(np.uint8), cmap='gray')
plt.show()



# plt.imshow(cv2.cvtColor(frizzy, cv2.COLOR_BGR2RGB))
# plt.show()
# plt.imshow(cv2.cvtColor(froomer, cv2.COLOR_BGR2RGB))
```
